```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools=len(school_data.loc[:,'school_name'])
total_students=len(student_data.loc[:,'student_name'])
total_budget=sum(school_data.loc[:,'budget'])
avg_math=student_data.loc[:,'math_score'].mean()
avg_reading=student_data.loc[:,'reading_score'].mean()
passing_math_filter=student_data.loc[student_data['math_score'] >= 70]['math_score'].count()
math_passing=(passing_math_filter/total_students)
passing_reading_filter=student_data.loc[student_data['reading_score']>=70]['reading_score'].count()
reading_passing=(passing_reading_filter/total_students)
overall_passing_filter=student_data[(student_data['math_score'] >= 70) & (student_data['reading_score']>=70)]['school_name'].count()
overall_passing=(overall_passing_filter/total_students)


data=({
"Total Schools": [total_schools],
"Total Students":[total_students],
"Total Budget":[total_budget],
"Avg Math Score":[avg_math],
"Avg Reading Score":[avg_reading],
"% Passing Math":[math_passing],
"% Passing Reading":[reading_passing],
"% Overall Passing":[overall_passing]})

district_info=pd.DataFrame(data)


district_info=district_info.style.format({"Total Students":"{:,}",
                          "Total Budget":"${:,}",
                           "Avg Math Score":"{:.1f}",
                           "Avg Reading Score":"{:.1f}",
                           "% Passing Math":"{:.1%}",
                           "% Passing Reading":"{:.1%}",
                           "% Overall Passing":"{:.1%}"})


district_info
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
by_school=school_data_complete.set_index('school_name').groupby(['school_name'])
school_type=by_school['type'].first()
total_stu=by_school['student_name'].size()
school_budget=by_school['budget'].first()
per_student_budget=(school_budget/total_stu)
avg_math_score=by_school['math_score'].mean()
avg_read_score=by_school['reading_score'].mean()
school_passing_math=school_data_complete[school_data_complete['math_score']>=70].groupby(['school_name']).size()
percent_passing_math=(school_passing_math/total_stu)
school_passing_reading=school_data_complete[school_data_complete['reading_score']>=70].groupby(['school_name']).size()
percent_passing_reading=(school_passing_reading/total_stu)
school_overall_passing=school_data_complete[(school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70)].groupby('school_name')['Student ID'].count()
percent_passing_overall=(school_overall_passing/total_stu)

school_sum={'School Type': school_type,
           'Total Students': total_stu,
           'School Budget': school_budget,
           'Per Student Budget': per_student_budget,
           'Average Math Score': avg_math_score,
           'Average Reading Score': avg_read_score,
           '% Passing Math': percent_passing_math,
           '% Passing Reading': percent_passing_reading,
           '% Overall Passing': percent_passing_overall
           }

school_sum=pd.DataFrame(school_sum)

school_sum_df=school_sum.style.format({"Total Students":"{:,}",
                          "School Budget":"${:,}",
                            "Per Student Budget":"${:,}",
                           "Average Math Score":"{:.1f}",
                           "Average Reading Score":"{:.1f}",
                           "% Passing Math":"{:.1%}",
                           "% Passing Reading":"{:.1%}",
                          "% Overall Passing":"{:.1%}"
                                   })

school_sum_df
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
top_ranked=school_sum.nlargest(20,'% Overall Passing')
top_ranked=top_ranked.style.format({"Total Students":"{:,}",
                          "School Budget":"${:,}",
                            "Per Student Budget":"${:,}",
                           "Average Math Score":"{:.1f}",
                           "Average Reading Score":"{:.1f}",
                           "% Passing Math":"{:.1%}",
                           "% Passing Reading":"{:.1%}",
                          "% Overall Passing":"{:.1%}"
                                   })
top_ranked
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
bottom_ranked=school_sum.nsmallest(5,'% Overall Passing')
bottom_ranked=bottom_ranked.style.format({"Total Students":"{:,}",
                          "School Budget":"${:,}",
                            "Per Student Budget":"${:,}",
                           "Average Math Score":"{:.1f}",
                           "Average Reading Score":"{:.1f}",
                           "% Passing Math":"{:.1%}",
                           "% Passing Reading":"{:.1%}",
                          "% Overall Passing":"{:.1%}"
                                   })
bottom_ranked
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#school_data_complete.columns
ninth_m=school_data_complete.loc[school_data_complete['grade'] == '9th'].groupby('school_name')['math_score'].mean()
tenth_m=school_data_complete.loc[school_data_complete['grade'] == '10th'].groupby('school_name')['math_score'].mean() 
eleventh_m=school_data_complete.loc[school_data_complete['grade'] == '11th'].groupby('school_name')['math_score'].mean()
twelfth_m=school_data_complete.loc[school_data_complete['grade'] == '12th'].groupby('school_name')['math_score'].mean() 

data=({'9th': ninth_m,
      '10th': tenth_m,
      '11th': eleventh_m,
      '12th': twelfth_m })
                                 
math_grade=pd.DataFrame(data)                                 
                                 
math_grade.style.format({'9th':"{:.2f}",
      '10th': "{:.2f}",
      '11th': "{:.2f}",
      '12th': "{:.2f}"})                       
                                 
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins=[0,584.999,614.999,644.999,9999]
bin_names = ["< $585", "$585 - 614", "$615 - 644", "> $644"]
school_sum['Budget Bin']=pd.cut(school_sum['Per Student Budget'], bins, labels=bin_names)

by_bin=school_sum.groupby('Budget Bin')['Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing'].mean()

by_bin.style.format({'Average Math Score': '{:.1f}', 
                              'Average Reading Score': '{:.1f}', 
                              '% Passing Math': '{:.1%}', 
                              '% Passing Reading':'{:.1%}', 
                              '% Overall Passing': '{:.1%}'})
```
